Найти выбросы по числовой переменной (использовать метод основанный на среднем и на межквартильном размахе). Объясните результат

Вывести статистику с выбросами и без. Сравните и объясните результат


```python
import matplotlib.pyplot as plt
df = df.dropna()
plt.boxplot(df["gdp_growth"])
plt.show() #Методом среднего
```

выбросы есть

```python
df["gdp_growth"].describe()
```

```python
mean = df["gdp_growth"].mean()
std_deviation = df["gdp_growth"].std()
threshold = 3 * std_deviation

# Идентифицируем выбросы
outliers = [x for x in df["gdp_growth"] if abs(x - mean) > threshold]

# Выводим результаты
print("Среднее значение данных:", mean)
print("Стандартное отклонение данных:", std_deviation)
print("Потенциальные выбросы:", outliers) #Методом среднего
```

```python
Q1 = df["gdp_growth"].quantile(0.25)
Q3 = df["gdp_growth"].quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers2 = [x1 for x1 in df["gdp_growth"] if ((x1 < lower) or (x1 > upper))]#При помощи межквартильного размаха

print("Потенциальные выбросы:", outliers2)
```

Вывод: после поиска выбросов двумя разными методами,мы учидели, что и в 1 и во 2 мтоде выбросами являются 27.96153807, 34.5, -36.65815267.

###Пример реализации

```python
import pandas as pd
import numpy as np


def corr_pearson(column_1, column_2):
  cov = np.cov(column_1, column_2)
  return cov/(column_1.std() * column_2.std())
```

```python
data = {
    'Рост': [170, 155, 180, 162, 175],
    'Вес': [65, 50, 75, 55, 70]
}
df = pd.DataFrame(data)
df
```

```python
corr_pearson(df["Рост"], df["Вес"])
```
